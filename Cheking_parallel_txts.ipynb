```python
#@title Загрузка библиотек
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
#@title Загрузка текстов
error_txt = open('error.txt').read()
right_txt = open('right.txt').read()
#дамп
from Frequency_dictionary import word_counts
```

```python
#@title Токены
words_tokens_right = word_tokenize(right_txt)
words_tokens_error = word_tokenize(error_txt)

sent_tokens_right = sent_tokenize(right_txt)
sent_tokens_error = sent_tokenize(error_txt)
```

```python
#@title Избавляемся от знаков препинания
import string
punct = set(string.punctuation)
punct.add('``')
punct.add('—')
punct.add("''" )
words_tokens_right = [x for x in words_tokens_right if x not in punct]
words_tokens_error = [x for x in words_tokens_error if x not in punct]
```

```python
#@title Статистика
len_w_r = len(words_tokens_right)
len_w_e = len(words_tokens_error)


len_s_r = len(sent_tokens_right)
len_s_e = len(sent_tokens_error)
```

```python
#@title Слова которых нет в дампе Википедии
right_lose = []
error_lose = []
```

```python
#@title Слова из еrror text, которых нет в right text
diff = []
```

```python
#@title Основная функция
for i in range(len_w_r):
  if words_tokens_error[i] != words_tokens_right[i]:
    diff.append([words_tokens_right[i], i, words_tokens_right[i][0]])
    if words_tokens_error[i] not in word_counts:
      error_lose.append([words_tokens_error[i], i, words_tokens_error[i][0]])
    else:
      if words_tokens_right[i] not in word_counts:
        right_lose.append([words_tokens_right[i], i, words_tokens_right[i][0]])
```

```python
#@title Вывод результатов
print('                          error text  |  rigt text')
print(f'Сравнение количества слов:        {len_w_e}  |  {len_w_r}')
print(f'Сравнение количества предложений:  {len_s_e}  |  {len_s_r}')
print('__________________________________________________________ ')
print(f'Ненайденные слова(1 - слово, 2 - порядковы индекс в тексте, 3 - первая буква): \n \n {error_lose}  |  {right_lose}')
print(f'Сравнение количества предложений:  {len_s_e}  |  {len_s_r}')
print('__________________________________________________________ ')
print(f'Исправленные слова(1 - слово, 2 - порядковы индекс в тексте, 3 - первая буква): \n \n {diff}')
```
